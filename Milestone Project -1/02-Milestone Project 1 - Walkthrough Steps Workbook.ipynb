**Step 1: Write a function that can print out a board. Set up your board as a list, where each index 1-9 corresponds with a number on a number pad, so you get a 3 by 3 board representation.**

```python
from IPython.display import clear_output

def display_board(board):
    print(board[7] +"|" + board[8] +"|"+ board[9])
    print("--"+"--"+"--")
    print(board[4] +"|" + board[5] +"|"+ board[6])
    print("--"+"--"+"--")
    print(board[1] +"|" + board[2] +"|"+ board[3])
```

**Step 2: Write a function that can take in a player input and assign their marker as 'X' or 'O'. Think about using *while* loops to continually ask until you get a correct answer.**

```python
def player_input(p):
    marker = ''
    while marker != "X" and marker != "O":
        marker = input("Player{0} please choose your mark: X or O".format(p)).upper()
    if (marker == "X" and p == 1) or (marker == "O" and p == 2):
        p1 = "X"
        p2 = "O"
    else:
        p1 = "O"
        p2 = "X"
    return (p1,p2)
```

**TEST Step 2:** run the function to make sure it returns the desired output

```python
player_input(1)
```

**Step 3: Write a function that takes in the board list object, a marker ('X' or 'O'), and a desired position (number 1-9) and assigns it to the board.**

```python
def place_marker(board, marker, position):
    while position not in range(1,10):
        position = int(input("please choose the position among 1-9"))
    board[position] = marker
```

**Step 4: Write a function that takes in a board and a mark (X or O) and then checks to see if that mark has won. **

```python
def win_check(board, mark, dict_Player):
    if board[1] == mark and board[2] == mark and board[3] == mark:
        print(f"{dict_Player[mark]} won")
    elif board[4] == mark and board[5] == mark and board[6] == mark:
        print(f"{dict_Player[mark]} won")
    elif board[7] == mark and board[8] == mark and board[9] == mark:
        print(f"{dict_Player[mark]} won")
    elif board[1] == mark and board[4] == mark and board[7] == mark:
        print(f"{dict_Player[mark]} won")
    elif board[2] == mark and board[5] == mark and board[8] == mark:
        print(f"{dict_Player[mark]} won")
    elif board[3] == mark and board[6] == mark and board[9] == mark:
        print(f"{dict_Player[mark]} won")
    elif board[1] == mark and board[5] == mark and board[9] == mark:
        print(f"{dict_Player[mark]} won")
    elif board[3] == mark and board[5] == mark and board[7] == mark:
        print(f"{dict_Player[mark]} won")
    else:
        return "game is on"
```

**Step 5: Write a function that uses the random module to randomly decide which player goes first. You may want to lookup random.randint() Return a string of which player went first.**

```python
import random

def choose_first():
    firstMove = random.randint(1,2)
    return 1 if firstMove == 1 else 2
choose_first()
```

**Step 6: Write a function that returns a boolean indicating whether a space on the board is freely available.**

```python
def space_check(board, position):
    return True if board[position] != "X" and board[position] != "O" else False
```

**Step 7: Write a function that checks if the board is full and returns a boolean value. True if full, False otherwise.**

```python
def full_board_check(board):
    for item in board:
        if item not in ["X", "O", "#"]:
            return False
    return True
test_board[8]=" "
display_board(test_board)
full_board_check(test_board)
```

**Step 8: Write a function that asks for a player's next position (as a number 1-9) and then uses the function from step 6 to check if it's a free position. If it is, then return the position for later use.**

```python
def player_choice(board):
    nextPosition = ''
    while nextPosition not in range(1,10):
        nextPosition = int(input("Please enter your next position (1-9): "))
    positionAvailability = space_check(board, nextPosition)
    if positionAvailability == False:
        print(f'position {nextPosition} is occupied')
    return nextPosition if positionAvailability else False
player_choice(test_board)
```

**Step 9: Write a function that asks the player if they want to play again and returns a boolean True if they do want to play again.**

```python
def replay():
    playAgain = ''
    while playAgain not in ['y', 'n']:
        playAgain = input("Play again? y or n: ").lower()
    if playAgain == 'y':
        return True
    else:
        return False
replay()
```

**Step 10: Here comes the hard part! Use while loops and the functions you've made to run the game!**

```python
gameon = True
while gameon:
    print('Welcome to Tic Tac Toe!\n')

    for index in range(1,10):
        test_board[index]=' '
        
    display_board(test_board)
    
    print("\nPlayer1 vs. Player2")
    print("Let's Toss for the Fisrt Move! Let's Go...\n")
    
    firstMove = choose_first()
    if firstMove == 1:
        print("Player1 goes first")
    else:
        print("Player2 goes first")
    [p1Mark, p2Mark] = player_input(firstMove)
    
    print(f"Player1 Mark: {p1Mark}") 
    print(f"Player2 Mark: {p2Mark}")
    dict_Player = {p1Mark:"Player1", p2Mark:"Player2"}
    
    fullBoardComplete = full_board_check(test_board)
    
    def letsPlay(mark):
        playerNextPosition = player_choice(test_board)
        
        while playerNextPosition == False:
            playerNextPosition = player_choice(test_board)
            
        place_marker(test_board, mark, playerNextPosition)
        display_board(test_board)
        
    moveCount = 1
    
    while not fullBoardComplete:
        if firstMove == 1:
            print("Player1 Turn")
            letsPlay(p1Mark)
            firstMove = 2
            fullBoardComplete = full_board_check(test_board)
        else:
            print("Player 2 Turn")
            letsPlay(p2Mark)
            
            firstMove = 1
            fullBoardComplete = full_board_check(test_board)
    
        if moveCount >= 5:
            for item in [p1Mark, p2Mark]:
                gameResult = win_check(test_board,item,dict_Player)
                if gameResult != "game is on":
                    fullBoardComplete = True
                    break
                elif moveCount == 9:
                    print("Match Draw")
                    break
    
        moveCount += 1
    gameon = replay()
    fullBoardComplete = False
```

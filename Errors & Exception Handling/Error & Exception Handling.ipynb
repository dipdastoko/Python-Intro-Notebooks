```python
try:
    result = 10 + 10
except:
    print("Hey, it's look like you are not adding correctly!")
else:
    print(result)
```

```python
try:
    f = open('testfile','r')
    f.write("Write a test line")
except TypeError:
    print("Hey there is type error!")
except OSError:
    print("Hey there is an OS error!")
finally:
    print("I always run")
```

```python
def ask_for_int():
    while True:
        try:
            result = int(input("please input a number: "))
        except:
            print("Whoops! That is not a number")
            continue
        else:
            print("Thank you")
            break
        finally:
            print("End of try/except/finally")
```

```python
ask_for_int()
```

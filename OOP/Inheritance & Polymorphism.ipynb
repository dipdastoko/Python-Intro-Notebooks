```python
class Animal():
    def __init__(self):
        print("Animal Created")
    def who_am_i(self):
        print("I am an animal.")
    def eat(self):
        print("I am eating")
```

```python
my_animal = Animal()
```

```python
my_animal.who_am_i()
```

```python
class Dog(Animal):
    def __init__(self):
        Animal.__init__(self)
        print("Dog Created")
```

```python
mydog = Dog()
```

```python
mydog.eat()
```

```python
class Dog():
    def __init__(self,name):
        self.name = name
    def speak(self):
        return self.name + " says WOOF!"
```

```python
class Cat():
    def __init__(self,name):
        self.name = name
    def speak(self):
        print(self.name + " says meow!")
```

```python
niko = Dog("niko")
```

```python
felix = Cat("felix")
```

```python
niko.name
```

```python
niko.speak()
```

```python
for pet in [niko, felix]:
    print(pet.speak())
```

```python
def pet_speak(pet):
    print(pet.speak())
```

```python
pet_speak(felix)
```

```python
class Animal():
    def __init__(self,name):
        self.name = name
    def speak(self):
        raise NotImplementedError("Subclass must implement this abstract method")
```

```python
myanimal = Animal('Fred')
```

```python
myanimal.speak()
```

```python
class Dog(Animal):
    def speak(self):
        return self.name + " says WOOF!"
```

```python
mydog = Dog('fido')
```

```python
mydog.speak()
```

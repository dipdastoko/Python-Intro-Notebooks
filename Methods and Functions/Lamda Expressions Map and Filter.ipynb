```python
def square(num):
    return num**2
```

```python
mylist = [1,2,3,4,5]
```

```python
for item in map(square,mylist):
    print(item)
```

```python
list(map(square,mylist))
```

```python
def splicer(mystring):
    if len(mystring)%2 == 0:
        return 'EVEN'
    else:
        return mystring[0]
```

```python
mynames = ['Sally', 'Andy', 'Goni']
```

```python
list(map(splicer, mynames))
```

```python
def check_even(num):
    return num%2==0
```

```python
mynums = [1,2,3,4,5,6]
```

```python
list(filter(check_even,mynums))
```

```python
list(map(check_even,mynums))
```

```python
list(filter(lambda num: num%2==0,mynums))
```

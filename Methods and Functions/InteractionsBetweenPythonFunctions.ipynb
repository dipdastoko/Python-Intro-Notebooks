```python
def player_guess():
    guess = ''
    while guess not in ['0','1','2']:
        guess = input("Pick a number: 0, 1 or 2")
    return int(guess)
```

```python
player_guess()
```

```python
def check_guess(mylist, guess):
    if mylist[guess]=='O':
        print("Correct!")
    else:
        print("Wrong!")
        print(mylist)
```

```python
#initial list
mylist = [' ','O',' ']
#shuffled list
mixedupList = shuffle_list(mylist)
#user guess
guess = player_guess()
#chekc guess
check_guess(mixedupList, guess)
```

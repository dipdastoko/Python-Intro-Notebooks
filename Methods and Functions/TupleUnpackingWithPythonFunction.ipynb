```python
work_hours = [('Abby',100),('Billy',200),('Cassie',300)]
```

```python
def employee_check(work_hours):
    current_max = 0
    employee_of_month = ''
    for name, wk in work_hours:
        if wk>current_max:
            current_max = wk
            employee_of_month = name
    return (employee_of_month, current_max)
```

```python
employee_check(work_hours)
```
